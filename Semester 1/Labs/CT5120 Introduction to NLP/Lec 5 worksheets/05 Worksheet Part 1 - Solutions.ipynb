# Parsing

Simple examples of drawing dependency parse trees and constituency parse tress

Some set up code


```python
%%capture
%tensorflow_version 1.x
!pip install benepar
import benepar
from benepar.spacy_plugin import BeneparComponent
import nltk
benepar.download('benepar_en3')
%matplotlib inline
from nltk.tree import Tree
from nltk.draw.tree import TreeView
import IPython
!apt install ghostscript python3-tk
### CREATE VIRTUAL DISPLAY ###
!apt-get install -y xvfb # Install X Virtual Frame Buffer
import os
os.system('Xvfb :1 -screen 0 1600x1200x16  &')    # create virtual display with size 1600x1200 and 16 bit color. Color can be changed to 24 or 8
os.environ['DISPLAY']=':1.0'
import spacy
nlp=spacy.load('en')
nlp.add_pipe(BeneparComponent('benepar_en3'))
```

```python
from spacy import displacy

def dependency_parse(text):
  displacy.render(nlp(text),jupyter=True)

dependency_parse("The girl, who is playing with cars, likes toast")
```

```python
def constituency_parse(text):
  tree = Tree.fromstring(list(nlp(text).sents)[0]._.parse_string)
  IPython.core.display.display(tree)

constituency_parse("this is a test")
```

Try parsing the following sentences

* Man bites dog
* They told you to support him
* The old man the boats

```python
dependency_parse("man bites dog")
```

```python
constituency_parse("man bites dog")
```

```python
dependency_parse("they told you to support him")
```

```python
constituency_parse("they told you to support him")
```

```python
constituency_parse("the old man the boats")
```
